```python
fig, axarr = plt.subplots(2, 1, figsize=(12, 8))


reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0])

reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1])
```

```python
fig,axarr=plt.subplots(2,2,figsize=(12,8))

reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0][0])

reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1][1])
```

```python
fig,axarr=plt.subplots(2,2,figsize=(12,8))

reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0][0],fontsize=12,color='mediumvioletred')
axarr[0][0].set_title('Points')

reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1][1])


reviews['variety'].value_counts().head(20).plot.bar(ax=axarr[0][1])

reviews['region_1'].value_counts().head(20).plot.bar(ax=axarr[1][0])

import seaborn as sns
sns.despine()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pokemon=pd.read_csv('Pokemon.csv')
```

```python
pokemon.head()
```

```python
fig,axarr=plt.subplots(2,1,figsize=(8,8))
```

```python
fig,axarr=plt.subplots(2,1,figsize=(8,8))

pokemon['Attack'].hist(ax=axarr[0])
axarr[0].set_title('Pokemon Attack Ratings')

pokemon['Defense'].hist(ax=axarr[1])
axarr[1].set_title('Pokemon Defense Ratings')
```
