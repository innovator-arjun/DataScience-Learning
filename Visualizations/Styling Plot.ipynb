```python
pokemon_data=pd.read_csv('pokemon.csv',index_col=0)
```

```python
pokemon_data.head()
```

```python
pokemon_data.plot.scatter(x='Attack',y='Defense',figsize=(12,6),title='Pokemon by Attack and Defense')
```

```python
print(pokemon_data.columns)
```

```python
ax=pokemon_data['Type 1'].value_counts().plot.bar(figsize=(20,8),
                                         
                                              fontsize=30
                                              )
ax.set_title('Pokemon by Primary Type'  ,fontsize=50)
```

```python
ax = pokemon_data['Total'].plot.hist(
    figsize=(12, 6),
    fontsize=14,
    bins=50,
    color='gray'
)
ax.set_title('Pokemon by Stat Total', fontsize=20)
```
